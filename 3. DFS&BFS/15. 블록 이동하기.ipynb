```python
# 그래프(이차원 리스트) 입력 받기
graph = eval(input())
# 그래프의 행,열 길이
n = len(graph)

# 최소 시간
mn = 1e9

# 움직이기
def move(now, direction):
    a, b = now[0][0], now[0][1]
    c, d = now[1][0], now[1][1]
    t = now[2] 

    next = now

    # 움직임
    ## 가로로 오른쪽
    if direction == 0:
        na, nb = a, b+1
        nc, nd = c, d+1
        nt = t+1
        if (0 <= nd <= n-1):
            if (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 가로로 아래쪽
    elif direction == 1:
        na, nb = a+1, b
        nc, nd = c+1, d
        nt = t+1
        if (0 <= na <= n-1):
            if (graph[na][nb] == 0) and (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 가로로 왼쪽
    elif direction == 2:
        na, nb = a, b-1
        nc, nd = c, d-1
        nt = t+1
        if (0 <= nb <= n-1):
            if (graph[na][nb] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 가로로 위쪽
    elif direction == 3:
        na, nb = a-1, b
        nc, nd = c-1, d
        nt = t+1
        if (0 <= na <= n-1):
            if (graph[na][nb] == 0) and (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 세로로 오른쪽
    elif direction == 4:
        na, nb = a, b+1
        nc, nd = c, d+1
        nt = t+1
        if (0 <= nc <= n-1):
            if (graph[na][nb] == 0) and (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 세로로 위쪽
    elif direction == 5:
        na, nb = a-1, b
        nc, nd = c-1, d
        nt = t+1
        if (0 <= na <= n-1):
            if (graph[na][nb] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 세로로 왼쪽
    elif direction == 6:
        na, nb = a, b-1
        nc, nd = c, d-1
        nt = t+1
        if (0 <= nc <= n-1):
            if (graph[na][nb] == 0) and (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 세로로 아래쪽
    elif direction == 7:
        na, nb = a+1, b
        nc, nd = c+1, d
        nt = t+1
        if (0 <= nc <= n-1):
            if (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]

    # 회전
    ## 3 -> 6
    elif direction == 8:
        na, nb = a, b
        nc, nd = c+1, d-1
        nt = t+1
        if (0 <= nd <= n-1):
            if (graph[nc][nd+1] == 0) and (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 6 -> 9
    elif direction == 9:
        na, nb = a, b
        nc, nd = c-1, d-1
        nt = t+1
        if (0 <= nd <= n-1):
            if (graph[nc+1][nd] == 0) and (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 9 -> 12
    elif direction == 10:
        na, nb = a-1, b+1
        nc, nd = c, d
        nt = t+1
        if (0 <= na <= n-1):
            if (graph[na][nb-1] == 0) and (graph[na][nb] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 12 -> 3
    elif direction == 11:
        na, nb = a+1, b+1
        nc, nd = c, d
        nt = t+1
        if (0 <= nb <= n-1):
            if (graph[na-1][nb] == 0) and (graph[na][nb] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 3 -> 12
    elif direction == 12:
        na, nb = a, b
        nc, nd = c-1, d-1
        nt = t+1
        if (0 <= nc <= n-1):
            if (graph[nc][nd+1] == 0) and (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 12 -> 9
    elif direction == 13:
        na, nb = a+1, b-1
        nc, nd = c, d
        nt = t+1
        if (0 <= nb <= n-1):
            if (graph[na-1][nb] == 0) and (graph[na][nb] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 9 -> 6
    elif direction == 14:
        na, nb = a+1, b+1
        nc, nd = c, d
        nt = t+1
        if (0 <= na <= n-1):
            if (graph[na][nb-1] == 0) and (graph[na][nb] == 0):
                next = [(na,nb), (nc,nd), nt]
    ## 6 -> 3
    elif direction == 15:
        na, nb = a, b
        nc, nd = c-1, d+1
        nt = t+1
        if (0 <= nd <= n-1):
            if (graph[nc+1][nd] == 0) and (graph[nc][nd] == 0):
                next = [(na,nb), (nc,nd), nt]
    
    return next

# dfs 선언
def dfs(now):
    global mn

    if not ((now[0][0] == n-1) and (now[0][1] == n-1)) or ((now[1][0] == n-1) and (now[1][1] == n-1)):
        for direction in range(16):
            if now != move(now, direction):
                next = move(now, direction)
                dfs(next)
    else:
        mn = min(mn, now[2])
    
# dfs 수행
dfs([(0,0), (0,0), 0])

# 최소 시간 출력
print(mn)
```
